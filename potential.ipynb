```python
import numpy as np
import matplotlib.pyplot as plt

rho_values = [20, 30, 40, 50]
e = -1.602176634e-19
ϵ = 8.8541878128e-12


for ρ in rho_values:
    # Define the potential function
    def V1(r):
        if r > ρ:
            return -e**2 / (4 * np.pi * ϵ * r)
        else:
            return float('nan')

    def V2(r):
        if 0 <= r <= ρ:
            return -(3 * e**2) * (1 / (8 * np.pi * ϵ * ρ)) * (1 - (r**2 / (3 * ρ**2)))
        else:
            return float('nan')

    # Generate a range of r values for each segment
    r1_values = np.arange(ρ, 100, 0.01)  # Adjust the end value as needed
    r2_values = np.arange(0.01, ρ, 0.01)

    # Calculate V(r) for each r
    V1_values = [V1(r) for r in r1_values]
    V2_values = [V2(r) for r in r2_values]

    # Create two separate plots
    plt.plot(r1_values, V1_values, label=f"r < ρ_{rho_values.index(ρ)}")
    plt.plot(r2_values, V2_values, label=f"r > ρ_{rho_values.index(ρ)}")

plt.xlabel(r"$r$")
plt.ylabel(r"$V(r)$")
plt.legend(prop={'size': 6}, loc=4)

plt.savefig('potential.png', dpi=500, bbox_inches='tight')
```
